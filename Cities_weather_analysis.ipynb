## Introduction of Dataset
This dataset provides **temperature data** for the top 100 cities worldwide as of `April 28th, 2024`. It consists of seven columns, each offering specific information to understand the weather conditions in different cities. The dataset includes the following columns:

### 1. `City`
The name of the city in English language. This column provides the official or commonly used name of each city. City names are essential for identifying the geographical location associated with the weather data.

### 2. `Temperature (¬∞C)`
The temperature recorded for each city on `April 28th, 2024`. Temperature is measured in degrees Celsius (`¬∞C`) and represents the average atmospheric temperature at the specified location on the given date. It is an essential parameter in weather analysis, indicating the warmth or coldness of the air.

### 3. `Wind Speed (m/s)`
The average wind speed measured at each location in `meters per second (m/s)`. Wind speed is a measure of the movement of air molecules and is an important factor in weather prediction and analysis.

### 4. `Latitude (¬∞)`
The latitude coordinates of each city, expressed in degrees. Latitude measures the north-south position of a location on the Earth's surface relative to the equator. Positive values represent locations in the northern hemisphere, while negative values represent locations in the southern hemisphere. Latitude ranges from `-90¬∞` (South Pole) to `+90¬∞` (North Pole), with `0¬∞` at the equator.

### 5. `Longitude (¬∞)`
The longitude coordinates of each city, expressed in degrees. Longitude measures the east-west position of a location on the Earth's surface relative to the Prime Meridian. Positive values represent locations east of the Prime Meridian, while negative values represent locations west of it. Longitude ranges from `-180¬∞` (West) to `+180¬∞` (East), with `0¬∞` at the Prime Meridian.

### 6. `Description`
This column provides details about the prevailing weather conditions in each city on `April 28th, 2024`. It includes descriptions such as `clear sky`, `scattered clouds`, `broken clouds`, `overcast clouds`, and `few clouds`, indicating the level of cloud cover and sky visibility at the time of observation.

### 7. `Country`
This column contains the name of the country to which each city belongs. It provides information about the city with its respective country.

#### 2. Use the OpenWeather API to collect real-time weather data for TOP 100 cities of the World.

````python
# first, we will load the list of cities from a CSV file

cities = pd.read_csv('top100cities_weather_data.csv')  
cities.tail()
````

````python
# Your API key
api_key = '65190f9772626f8c7f3e0b041a41dd07'

# Define an empty list to store the results
weather_data = []

# Loop through each city and its corresponding country in the DataFrame
for index, row in cities.iterrows():
    city = row['City']
    country = row['Country']  
    # Construct the API request URL
    url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric'

    # Make the request to the API
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()  # Convert response to JSON format

        # Extract weather data
        temp = data['main']['temp']
        feels_like = data['main']['feels_like']
        temp_min = data['main']['temp_min']
        temp_max = data['main']['temp_max']
        humidity = data['main']['humidity']
        pressure = data['main']['pressure']
        wind_speed = data['wind']['speed']
        desc = data['weather'][0]['description']
        #country = data['sys']['country']
        clouds = data['clouds']['all']
        lon = data['coord']['lon']
        lat = data['coord']['lat']
        sunrise_unix = data['sys']['sunrise']
        sunset_unix = data['sys']['sunset']

        # Convert Unix timestamps to human-readable format
        sunrise = datetime.fromtimestamp(sunrise_unix).strftime('%I:%M %p')
        sunset = datetime.fromtimestamp(sunset_unix).strftime('%I:%M %p')

        # Append the data to the list
        weather_data.append({
            'Country': country,
            'City': city,
            'Temperature': temp,
            'Feels_Like': feels_like,
            'Min_Temp': temp_min,
            'Max_Temp': temp_max,
            'Humidity': humidity,
            'Pressure': pressure,
            'Wind_Speed': wind_speed,
            'Cloudiness': clouds,
            'Description': desc.capitalize(),
            'Longitude': lon,
            'Latitude': lat,
            'Sunrise': sunrise,
            'Sunset': sunset
        })

    else:
        # If the request fails, append an error message for that city
        weather_data.append({
            'City': city,
            'Error': f'Error fetching data: {response.status_code}'
        })

# Convert the list of weather data into a pandas DataFrame
weather_df = pd.DataFrame(weather_data)
````

````python
# save the dataset that is fetched from the API to a CSV file:

weather_df.to_csv('top_100_cities_of_the_world_weather_data.csv', index=False)
````

> BUT, there is a difference between `str` and `category` data type, let's see the difference first:

In Python/Pandas, `str` and `category` are two different data types used to handle textual data, each with its own characteristics and use cases. Here‚Äôs a detailed comparison:

---
* `str` Data Type
---

1. **Definition**:
   - Represents strings of text.
   - Standard Python `str` type, used to store textual data.

2. **Storage**:
   - Each unique string is stored as is.
   - Memory usage can be higher, especially with a large number of unique strings.

3. **Operations**:
   - Supports all standard string operations and methods.
   - Slower for operations involving large datasets with many unique strings, due to the lack of optimization for categorical data.

4. **Use Case**:
   - Ideal for columns with a small number of unique strings or when the exact text value is important.

5. **Example**:
   ```python
   df['text_column'] = df['text_column'].astype(str)
   ```
---
* `category` Data Type
---

1. **Definition**:
   - A Pandas-specific data type optimized for categorical data.
   - Represents a fixed number of possible values (categories).

2. **Storage**:
   - Internally stores data as integer codes and a separate array of category labels.
   - More memory-efficient, especially with many repeated values.

3. **Operations**:
   - Limited string operations compared to `str`.
   - Faster for operations involving large datasets with repeating categories due to the use of integer codes.
   - Supports category-specific operations like reordering, adding, or removing categories.

4. **Use Case**:
   - Ideal for columns with a limited set of distinct values that repeat often (e.g., gender, color, or category labels).
   - Useful for improving performance and memory usage in large datasets with repetitive text.

5. **Example**:
   ```python
   df['category_column'] = df['category_column'].astype('category')
   ```

**Summary**

- Use `str` when you need full string functionality and when dealing with text data where every unique string matters.
- Use `category` when you have a column with a limited number of unique values that repeat often, as it provides efficiency improvements in both memory and performance. 

**Choosing the right type depends on the specific needs of your data and the operations you need to perform.**
